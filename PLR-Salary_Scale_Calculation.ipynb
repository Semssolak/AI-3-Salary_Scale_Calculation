Polynomial Linear Regression General Formula:

y = a + b1*x + b2*x^2 + b3*x^3 + b4*x^4 + ....... + bN*x^N


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


# We take our data set with the help of pandas and transfer it into our dataframe object, df.
df = pd.read_csv("polynomial.csv",sep = ";")
```

```python
df
```

```python
# Let's take a look at our dataset
plt.scatter(df['Experience'],df['Salary'])
plt.xlabel('Experience (year)')
plt.ylabel('Salary')
plt.savefig('1.png', dpi=300)
plt.show()
```

```python
# As can be seen, the data is not distributed in a linear structure
# If we apply linear regression to this data set, we see an inappropriate prediction line:

reg = LinearRegression()
reg.fit(df[['Experience']],df['Salary'])

plt.xlabel('Experience (year)')
plt.ylabel('Salary')

plt.scatter(df['Experience'],df['Salary'])   

xaxis = df['Experience']
yaxis = reg.predict(df[['Experience']])
plt.plot(xaxis, yaxis,color= "green", label = "linear regression")
plt.legend()
plt.show()
```

#### It is very bad for prediction. It is not right to apply linear regression for this data set.

### We decided that polynomial regression, one of the regression types, should be applied for this data set. Now let's see how we implement it:

We adapt our x value to fit the polynomial function above

So => 1, x, x^2 (N=2)

```python
# we call the PolynomialFeatures function to create a polynomial regression object
# When calling this function, we specify the degree (N) of the polynomial:
polynomial_regression = PolynomialFeatures(degree = 4)

x_polynomial = polynomial_regression.fit_transform(df[['Experience']])
```

```python
# Creates our reg object, which is our regression model object, and calls its fit method to fit the x_polynomial and y axes
# so we train our regression model with real data available:
reg = LinearRegression()
reg.fit(x_polynomial,df['Salary'])
```

### Now that our model is ready and trained, let's see what kind of result graph our model creates based on the data at hand:

```python
y_head = reg.predict(x_polynomial)
plt.plot(df['Experience'],y_head,color= "red",label = "polynomial regression")
#plt.plot(xekseni, yekseni,color= "green", label = "linear regression")
plt.legend()

#Let's scatter our data set as points and see if polynomial regression fits:
plt.scatter(df['Experience'],df['Salary'])   

plt.show()
```
